# График суммарной численности населения в разбивке по регионам.

```python
df1 = pd.read_csv("countries_of_the world.csv")
df1.info()
```

```python
df2 = pd.DataFrame()
```

```python
df2 = df1.groupby("Region")["Population"].sum().to_frame().reset_index()
```

```python
df2
```

```python
import matplotlib.pyplot as plt
plt.scatter(df2["Population"],df2["Region"],label = "PopulationByRegion")
plt.legend(loc = "upper right")
plt.show()
```

# в каком регионе наблюдается наибольший миграционный отток населения.

```python
df4 = pd.read_csv("countries_of_the world.csv")
df4.info()
```

```python
df4.isnull().sum()
df4.dropna(subset = ["Net migration"], inplace = True )
df4.isnull().sum()
```

```python
df4['Net migration'] = df4['Net migration'].replace(',','.', regex=True).astype(float)
df4["NEW"] = df4["Net migration"]*df4["Population"]/1000
```

```python
pd.options.display.float_format = '{:,.1f}'.format
df4.groupby("Region")["NEW"].sum().to_frame().reset_index()
```
